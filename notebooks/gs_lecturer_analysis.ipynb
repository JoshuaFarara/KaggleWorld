# Georgia Southern University Lecturer & Professor Salary Analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio

# pd.describe_option()
# pd.set_option('display.width', 1000)
pd.set_option("display.max_rows", 999)
# pd.set_option("max_colwidth", 200)
```

```python
df_lecturer = pd.read_csv("/content/Lecturer_SalaryTravel.csv")
df_professor = pd.read_csv("/content/Professor_SalaryTravel.csv")
```

## Lecturer Analysis

```python
df_lecturer.head()

```

```python
df_lecturer.info()
```

```python
# @title Top 10 Highest Paid Lecturers

import matplotlib.pyplot as plt

df_sorted = df_lecturer.sort_values('Salary', ascending=False).head(10)
plt.bar(df_sorted['Name'], df_sorted['Salary'])
plt.xlabel('Name')
plt.ylabel('Salary')
plt.xticks(rotation=90)
_ = plt.title('Top 10 Highest Paid Lecturers')
```

```python
# @title 10 Lowest Paid Lecturers

import matplotlib.pyplot as plt

df_sorted = df_lecturer.sort_values('Salary', ascending=True).head(10)
plt.bar(df_sorted['Name'], df_sorted['Salary'])
plt.xlabel('Name')
plt.ylabel('Salary')
plt.xticks(rotation=90)
_ = plt.title('10 Lowest Paid Lecturers')

# Adding text annotations
for i, value in enumerate(df_sorted['Salary']):
    plt.text(i, value + 500, str(value), ha='center', va='bottom')

plt.show
```
